### Algebra liniowa

Suma cyfr pewnej liczby trzycyfrowej wynosi 18. Cyfra dziesiątek jest o 1 większa od cyfry jedności. Jeśli zamienimy miejscami cyfrę setek i dziesiątek, to otrzymamy liczbę o 180 większą od początkowej. Wyznacz liczbę początkową.

$$\left\{\begin{array}{rcl}
x + y + z & = & 18\\
y - z     & = & 1\\
100x + 10y + z + 180 & = & 100y + 10x + z
\end{array} \right.$$

$$\left\{\begin{array}{rcl}
x + y + z & = & 18\\
y - z     & = & 1\\
90x - 90y & = & -180
\end{array} \right.$$

```python
from scipy import linalg
import numpy as np

# zapis macierzowy
a = np.array([[1, 1, 1], [0, 1, -1], [90, -90, 0]])
b = np.array([18, 1, -180])

x = linalg.solve(a, b)

print(x)
```

### Kombinatoryka

Ile haseł ośmioznakowych można utworzyć używając małych i wielkich liter oraz cyfr przy założeniu, że się nie powtarzają.


```python
from scipy.special import perm

res = perm(26 + 26 + 10, 8)
print(res)
```

### Optymalizacja

Obliczenia numeryczne np. wyszukiwanie miejsc zerowych funkcji.

```python
from scipy.optimize import root
import numpy as np


def func(x):
    return 2 * x - 2


sol = root(func, 0.1)
print(sol)
```

```python
# z wykorzystaniem wyrażenia lambda
sol = root(lambda x: 2 * x - 2, 0.1)
print(sol)
```

### Algorytmy grafowe

Możliwe jest np. wyznaczanie najkrótszych ścieżek w grafie.

```python
from scipy.sparse.csgraph import shortest_path
import numpy as np

# macierz sąsiedztwa z odległościami
graph = np.array([[0, 100, 90], [100, 0, 9], [90, 9, 0]])
print(graph)
shortest_path(graph)  # macierz z najkrótszymi odległościami
```

## SymPy

[SymPy](https://www.sympy.org) to biblioteką do obsługi obliczeń symbolicznych.

```python
!pip install sympy
```

```python
from sympy import Symbol, factor

x = Symbol("x")
y = Symbol("y")
print(y + x + x + 1 + x * x - y)  # uproszczenie wyrażenia
print(factor(x + x + 1 + x * x))  # zamiana na iloczyn
print(factor(x * y + y * x + 1 + x**2 * y**2))
```

## Co dalej?

W analizie danych przydatne mogą być:
- [JAX](https://jax.dev) - interfejs podobny do NumPy do obliczeń wykonywanych na procesorach CPU, GPU lub TPU, w środowiskach lokalnych lub rozproszonych
- [statsmodels](https://www.statsmodels.org) - skupia się na modelowaniu statystycznym i testach hipotez; analiza szeregów czasowych, regresji liniowej lub gdy potrzebujemy szczegółowych podsumowań statystycznych
- [Dask](https://www.dask.org) - równoległe wykonywanie operacji na wielu rdzeniach procesora lub w klastrach; umożliwia pracę z danymi niebieszczącymi się w pamięci RAM
- [DuckDB](https://duckdb.org) - zapytania SQL bezpośrednio na plikach Parquet, CSV czy JSON bez ich wcześniejszego importowania; DuckDB i Polars używają Apache Arrow, więc współdzielą ten sam obszar pamięci, co pozwala na pracę na danych bez ich kopiowania
- [Daft](https://www.daft.ai) - zaprojektowana do pracy z danymi nieustrukturyzowanymi jak obrazy, wideo, dźwięk oraz do pracy rozproszonej
- [Ibis](https://ibis-project.org) - umożliwa uruchamianie wspólnego kodu na różnych silnikach przetwarzania danych typu DuckDB, Polars, BigQuery, Snowflake czy PostgreSQL
- [Great Expectations](https://greatexpectations.io) - testy jednostkowe na danych, zapewnianie jakości danych
- [Pydantic](https://pydantic.dev) - walidacja danych wejściowych
- [FastAPI](https://fastapi.tiangolo.com) - budowanie webowego API do udostępniania danych
